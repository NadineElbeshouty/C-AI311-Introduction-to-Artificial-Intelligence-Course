# Task 1: Convert Lowercase Characters in a given String to Uppercase

### Objective:
The objective of this task is to create a Python class called Newstring. This class is designed for manipulating and analyzing strings, specifically converting lowercase characters to uppercase.

### Task Details:

* Create the Newstring Class:

* Define a class called Newstring with the following attributes and methods:

### Attributes and Methods:
* nstr (str): The original string to be stored and manipulated.
* length (int): The length of the original string.
* uppercase(self): Converts all lowercase characters in the string to uppercase and returns the modified string.

````python
somestr = 'hEllo mR. aHmed !'
newstrlist = []
for letter in somestr:

    if 97 <= ord(letter) <= 122:

        newstrlist.append(chr(ord(letter) - 32))
    else:
        newstrlist.append(letter)
newstrlist
''.join(newstrlist)
````

````python
class Newstring:
    """
    A class for manipulating and analyzing strings.
    Attributes:
        nstr (str): The original string.
        length (int): The length of the original string.
    Methods:
        uppercase(): Converts all lowercase characters in the string to uppercase.
    """

    def __init__(self, nstr):
        """
        Initializes a Newstring object with the provided string.
        """
        self.length = len(nstr)
        self.nstr = nstr

    def uppercase(self):
        """
        Converts all lowercase characters in the string to uppercase.
        """
        somestr = self.nstr
        newstrlist = []
        for string in somestr:
            if 97 <= ord(string) <= 122:
                newstrlist.append(chr(ord(string) - 32))
            else:
                newstrlist.append(string)
        return ''.join(newstrlist)
````

````python
newstr = Newstring('string')
````

````python
print(newstr.length)
````

````python
print(newstr.uppercase())
````

## Recap

````python
class MyClass:

    class_variable = [1, 2, 3]

    def __init__(self):
        print('I am the constructor!')
        self.variable = 'I am an object variable'
        self._variable = 'I am a private object variable (by convention xD)'

    def my_method(self):
        print('I am a method!')

    def my_returning_method(self, something):
        return something

    def my_variable_returning_method(self):
        return self._variable

    def _my_private_method(self):
        print('I am a private method and I "should" not be called from outside the class')

    @classmethod
    def my_class_method(cls):
        print('I am a class method and I implicitly get my class as my first argument:', cls)

    @staticmethod
    def my_static_method():
        print('I am a class method and I behave just like any normal method, just call me from my class!')
````

Class method vs Static Method

* A class method takes cls as the first parameter while a static method needs no specific parameters.
* A class method can access or modify the class state while a static method cant access or modify it.
* In general, static methods know nothing about the class state. They are utility-type methods that take some parameters and work upon those parameters
* We use `@classmethod` decorator in python to create a class method and we use `@staticmethod` decorator to create a static method in python.

## Requirement

Now let's use what we learned so far to *estimate $\pi$*!

Consider a dart board shaped like a circle enclosed by a square and you are throwing darts at it. You are very good and you never land a dart off the square but you are not aiming for anything in particular inside it, so a dart has an equal chance of falling anywhere inside the square!

````python
from matplotlib import pyplot as plt
from matplotlib.patches import Circle, Rectangle

ax = plt.axes()
ax.add_patch(Circle((0,0), 1, fill=False, edgecolor='blue', aa=True, lw=5))
ax.add_patch(Rectangle((-1, -1), 2, 2, fill=False, edgecolor='red', aa=True, lw=5))

ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.axis('equal')
ax.grid(False)
ax.set_axis_off()
````

Then, the chance of a dart to fall inside the circle is the ratio of the circle area to the square area. $$p(circle) = \dfrac{circle\ area}{square\ area} = \dfrac{\pi r^2}{(2r)^2} = \dfrac{\pi}{4}$$ This means that $$\pi = 4 * p(circle)$$

Now, the remaining question is, how do we estimate the probability? We can do that by simulating random thrown darts, counting the one ones that land inside the circle and divide that by the total number of thrown darts!

$$p(circle) = \dfrac{number\ of\ darts\ inside\ the\ circle}{total\ number\ of\ darts}$$

That way of estimating probabilites is called *Monte Carlo* simulation.

````python
from matplotlib import pyplot as plt
from matplotlib.patches import Circle, Rectangle
import random

ax = plt.axes()
ax.add_patch(Circle((0,0), 1, fill=False, edgecolor='blue', aa=True, lw=5))
ax.add_patch(Rectangle((-1, -1), 2, 2, fill=False, edgecolor='red', aa=True, lw=5))

x = [random.uniform(-1, 1) for _ in range(1000)]
y = [random.uniform(-1, 1) for _ in range(1000)]
ax.scatter(x, y, s=5)

ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.axis('equal')
ax.grid(False)
ax.set_axis_off()
````
